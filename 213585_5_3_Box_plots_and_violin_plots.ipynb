```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Ящики с усами (box plots)** и **виолончельные графики (violin plots)** используются для визуализации распределения данных.
* **Box plot** отображает медиану, квартили и выбросы.
* **Виолончельный график** сочетает в себе информацию о плотности распределения и элементы ящика с усами.

Для этого урока будем использовать датасет tips.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Box Plot

[Ящики с усами](https://seaborn.pydata.org/generated/seaborn.boxplot.html) хорошо подходят для анализа распределений – поиска квартилей и выявления выбросов. В seaborn можно построить боксплот с помощью функции **seaborn.boxplot()**.

В следующем примере исследуем распределение чаевых по дням недели *(x='day' и y='tips')* и применим более читаемую палитру с помощью *palette='pastel'*:

```python
sns.boxplot(data=tips, x='day', y='tip', palette='pastel')

plt.title('Размер чаевых в зависимости от дня недели')
plt.xlabel('День недели')
plt.ylabel('Размер чаевых')

plt.show()
```

**Как читать Boxplot?**

![boxplot.png](https://external-content.duckduckgo.com/iu/?u=http%3A%2F%2Fdrive.google.com/uc?id=1b5VW98kZm562tiP3m2J7AaTT9AFwkJft&f=1)

Выбросы определяются по алгоритму с 1,5 * IQR.

А теперь с помощью параметра *hue=* посмотрим, как будет выглядеть этот график, но с дополнительным разделением на пол.

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex', palette='Set2')

plt.title('Размер чаевых по дням недели с разделением по полу')
plt.xlabel('День недели')
plt.ylabel('Размер чаевых')
plt.legend(title='Пол')

plt.show()
```

## Виолончельный график

[Виолончельный график](https://seaborn.pydata.org/generated/seaborn.violinplot.html) совмещает в себе свойства боксплота и графиков плотности распределения. С его помощью можно получить больше информации о характере распределения данных.

В следующем примере повторим тот график, который делали в прошлый раз, но вместо boxplot'а построим виолончельный график с помощью **seaborn.violinplot()**

```python
sns.violinplot(data=tips, x='day', y='tip', palette='muted')

plt.title('Распределение суммы счета по дням недели')
plt.xlabel('День недели')
plt.ylabel('Сумма счета')

plt.show()
```

А теперь с помощью трижды изученного нами *hue=* посмотрим на распределение по мужчинам и женщинам. Но тут все немного интереснее – появляется аргумент *split=*. При значении *True* вместо того, чтобы рисовать два отдельных violin plot'а для мужчин и женщин, один график просто разделится на две части. Выглядит очень интересно!

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='Set3')

plt.title('Распределение суммы счета по дням недели с разделением по полу')
plt.xlabel('День недели')
plt.ylabel('Сумма счета')

plt.show()
```
