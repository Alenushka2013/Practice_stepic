Разведочный анализ данных мы будем проводить на известном датасете о пассажирах Титаника.

Анализ данных состоит из следующих этапов:

1.Определение проблемы: если анализ данных, Big Data, машинное обучение, прогнозная аналитика, бизнес-аналитика или любое другое модное слово является решением, то в чем проблема?

2.Сбор данных: Джон Нейсбитт в своей книге Megatrends 1984 написал, что мы «тонем в данных, но стремимся к знаниям». Так что, скорее всего, наборы данных уже существуют где-то в каком-то формате. Может быть внешним или внутренним, структурированным или неструктурированным, статическим или потоковым, объективным или субъективным и т. д. Вам не нужно изобретать велосипед, вам просто нужно знать, где его найти. На следующем этапе мы будем беспокоиться о преобразовании «грязных данных» в «чистые данные».

3.Подготовка данных к использованию: этот шаг часто называют обработкой данных, необходимым процессом для превращения «диких» данных в «управляемые». Преодоление данных включает в себя реализацию архитектур данных для хранения и обработки, разработку стандартов управления данными для обеспечения качества и контроля, извлечение данных и очистку данных для выявления отсутствующих или выпадающих точек данных.

4. Исследовательский анализ: любой, кто когда-либо работал с данными, знает, что такое «мусор на входе», «мусор на выходе» (GIGO). Поэтому важно использовать описательную и графическую статистику для поиска потенциальных проблем, закономерностей, классификаций, корреляций и сравнений в наборе данных. Кроме того, категоризация данных (то есть качественная и количественная) также важна для понимания и выбора правильной проверки гипотез или модели данных.

5. Данные модели. Подобно описательной и логической статистике, моделирование данных может либо суммировать данные, либо предсказывать будущие результаты. Ваш набор данных и ожидаемые результаты определят доступные для использования алгоритмы. Важно помнить, что алгоритмы - это инструменты, а не волшебные палочки или серебряные пули. Вы все еще должны быть мастером своего дела, который знает, как выбрать правильный инструмент для работы. Неправильная модель может в лучшем случае привести к плохой работе, а в худшем - к неправильному выводу (который используется в качестве практического интеллекта).

6. Проверка и реализация модели данных: после того, как вы обучили свою модель на основе подмножества ваших данных, пришло время протестировать вашу модель. Это помогает гарантировать, что вы не переобучаете свою модель или не сделали ее настолько специфичной для выбранного подмножества, что она не будет точно соответствовать другому подмножеству из того же набора данных. На этом этапе мы определяем, соответствует ли наша модель нашему набору данных, обобщает или недостаточно.

7. Оптимизация и разработка стратегии: это этап, на котором вы повторяете процесс, чтобы сделать его лучше ... сильнее ... быстрее, чем было раньше.

```python
import pandas as pd

# Читаем данные
df = pd.read_csv('2008.csv')

# Фильтруем только задержки вылета (DepDelay > 0)
delayed_flights = df[df['DepDelay'] > 0]

# 1. Найдем топ-5 рейсов с наибольшей общей задержкой отправления
top_5_flights = delayed_flights.groupby('FlightNum')['DepDelay'].sum().nlargest(5).index

# 2. Подсчитаем количество случаев задержки для этих рейсов
delay_counts = df[df['FlightNum'].isin(top_5_flights) & (df['DepDelay'] > 0)].groupby('FlightNum').size()

# 3. Подсчитаем количество задержек, вызванных погодой
weather_delays = df[df['FlightNum'].isin(top_5_flights) & (df['WeatherDelay'] > 0)].groupby('FlightNum').size()

# Вывод результатов
print("Топ-5 рейсов с наибольшей суммарной задержкой отправления:", top_5_flights.tolist())
print("\nКоличество случаев задержек отправления для этих рейсов:")
print(delay_counts)
print("\nКоличество задержек, вызванных погодными условиями для этих рейсов:")
print(weather_delays)
```
