```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Диаграммы рассеивания (scatter plots) и линейные графики (linear plots)** – это основныe инструментs для визуализации зависимости между двумя переменными.
* **Диаграмма рассеивания** представляет собой плоскость, на которой каждая точка представляет значения двух переменных. Можно сравнить это с графиком функции
* С помощью **линейных графиков** можно визуализировать тренды на графике, где каждая точка соединяется линией

Для сегодняшнего урока будем использовать тот же самый датасет, что и в прошлый раз.

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

## Scatterplot

[Диаграмма рассеивания](https://seaborn.pydata.org/generated/seaborn.scatterplot.html) позволяет выявлять тренды, кластеры, а так же аномальные значения. В seaborn можно построить с помощью **seaborn.scatterplot()**.

Исследуем взаимосвязь между длиной клюва и длиной плавников у пингвина по видам. Каждая точка здесь – определенный пингвин, а его значения по X и Y – характеристики его клюва и плавников.

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species')

plt.title('Длина клюва и плавников у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавников (мм)')
plt.legend(title='Вид пингвина')


plt.show()
```

Теперь, немного усложним наш график – сделаем размер наших точек зависимым от размера пингвина с помощью аргумента **size=**, в который мы передадим признак, по которому будет меняться размер точек. С помощью **sizes=** укажем интервал размера наших точек.

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', size='body_mass_g',
                 hue='species', sizes=(20, 200))

plt.title('Диаграмма рассеивания: Взаимосвязь длины клюва и плавников с массой тела')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')


plt.show()
```

Легенда графика перекрывает большую его часть. С помощью **legend=false** уберем ее.

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', size='body_mass_g',
                 hue='species', palette='deep', sizes=(20, 200), legend=False)

plt.title('Диаграмма рассеивания: Взаимосвязь длины клюва и плавников с массой тела')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')

plt.show()
```

## Линейные графики

[Линейные графики](https://seaborn.pydata.org/generated/seaborn.lmplot.html) позволяют находить тренды в данных. С помощью них так же можно следить за изменением переменной во времени (но это не наш сегодняшний вариант).

Можно построить с помощью **seaborn.lmplot()**

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm')

plt.title('Длина клюва и плавников у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')

plt.show()
```

В этом графике так же присутствует аргумент **hue=**. Давайте посмотрим на тренды у разных видов пингвинов.

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm',
           hue='species', legend_out=False)


plt.title('Длина клюва и плавников у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')
plt.legend(title='Виды пингвинов')

plt.show()
```

С помощью **sns.lmplot()** можно строить графики для различных типов регресcии, а так же включать различные типы доверительных интервалов для более точного анализа данных.


```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', order=2, ci=None)

plt.title('Взаимосвязь длины клюва и плавников')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')

plt.show()
```

В примере выше с помощью аргумента *order=2* мы указали, что нужно построить
график полиномиальной регрессии второго порядка. Это может быть полезно для нелинейных зависимостей.
