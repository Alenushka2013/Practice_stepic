```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
flights = sns.load_dataset("flights")
```

В этом уроке разберем парные графики (pair plots) и матричные графики (matrix plots). Они помогают визуализировать данные в многомерных наборах.

## Парные графики

[Парный график (pair plot)](https://seaborn.pydata.org/generated/seaborn.pairplot.html) – это график, который позволяет отобразить попарные отношения между различными парами переменных в наборе данных. Парные графики часто используются для идентификации зависимостей между переменными и иногда используются для классификации точек данных. Из себя представляет массив из диаграмм рассеяния и гистограмм для каждой пары числовых переменных в наборе данных. Построить такой график можно с помощью **seaborn.pairplot()**

Построим такой график для набора данных Iris с разделением на группы по видам:

```python
sns.pairplot(iris, hue='species', palette='husl',
              diag_kind='kde', markers=['o', 's', 'D'], plot_kws={'alpha': 0.7})

plt.suptitle("Парный график для набора данных Iris", y=1.02)
plt.show()
```

В графике выше с помощью **diag_kind='kde'** на главной диагонали мы заменили гистограммы на графики плотности распределения, а при помощи **markers=** добавили уникальные маркеры для каждой из категории (поменяли форму, короче). **plot_kws={'alpha': 0.7 }** устанавливает прозрачность точек для улучшения визуализации при наложении.

## Матричные графики

[Матричные графики](https://seaborn.pydata.org/generated/seaborn.heatmap.html) включают тепловые карты (как ту, которую мы разбирали в прошлом уроке) и отображают данные в формате, где строки и столбцы представляют разные категории или параметры.

Построим график частоты пассажиров по разным месяцам разных лет из датасета flights:

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)

plt.title('Тепловая карта пассажиропотока по месяцам и годам')
plt.xlabel('Год')
plt.ylabel('Месяц')

plt.show()
```

**Что мы сделали?**
1. Сначала мы создали сводную таблицу с индексом в виде месяцев, а колонками в виде лет.
2. Потом запихнули эту таблицу в *data=* функции **seaborn.heatmap()**. Вдобавок к этому сказали сиборну писать наши значения *(annot=True)* в целочисленном формате *(fmt='d')*, передали цветовую палитру *(cmap="YlGnBu")* и указали размер линий между ячейками (и то что они вообще есть) с помощью *linewidth=.5*.
3. Ну и по традиции подписали название графика и его осей.
