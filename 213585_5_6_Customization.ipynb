```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Seaborn предлагает множество инструментов для настройки стилей и цветовых палитр, позволяя делать визуализацию данных не только информативной, но и эстетически привлекательной. В этом уроке мы научимся изменять стили графиков и выбирать палитры для улучшения восприятия данных.

```python
tips = sns.load_dataset('tips')
```

## Стили графиков в Seaborn

Seaborn предоставляет четыре основных стиля: **whitegrid, darkgrid, white и dark**. Они помогают задать общую атмосфер графика. Ставятся с помощью **seaborn.set_style('style_name')**

* *whitegrid* — стиль с белым фоном и сеткой (подходит для аналитических визуализаций).
* *darkgrid* — стиль с темным фоном и сеткой (универсальный, подходит для большинства графиков).
* *white* — стиль с белым фоном без сетки (минималистичный).
* *dark* — стиль с темным фоном без сетки (подходит для презентаций и отчетов).

Поочередно попробуем разные стили на одном и том же боксплоте, чтобы увидеть разницу.

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("whitegrid")
plt.show()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("darkgrid")
plt.show()
```

```python
sns.set_style("white")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("white")
plt.show()
```

```python
sns.set_style("dark")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("dark")
plt.show()
```

## Цветовые палитры

Цветовая палитра — это основа выразительности графика. Seaborn предлагает несколько встроенных палитр, таких как deep, muted, bright, pastel, dark, и colorblind, а также позволяет создавать палитры вручную. Выставить палитру можно через **seaborn.set_palette('palette_name')**

Основные типы цветовых палитр:
* **Qualitative** — для категорий (например, deep, bright, pastel).
* **Sequential** — для непрерывных данных, где оттенки изменяются от светлого к темному (например, Blues, Greens).
* **Diverging** — для данных с положительными и отрицательными значениями (например, coolwarm, RdBu).

```python
# Построим на палитру для категорий
sns.set_palette("bright")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='tip', y="total_bill", hue="sex", alpha=0.7)
plt.legend(title="Пол")
plt.show()
```

## Настройка темы

Помимо стиля и палитры, в Seaborn можно настроить такие параметры темы, как размер шрифта, ширина линий и отступы, используя функцию **seaborn.set_context()**. Эта функция настраивает параметры отображения под конкретные условия (например, отчеты или презентации).

```python
# Установим контекст для презентации
sns.set_context("talk")

plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="muted")
plt.title("Сумма счета по дням недели (контекст презентации)")
plt.show()
```
