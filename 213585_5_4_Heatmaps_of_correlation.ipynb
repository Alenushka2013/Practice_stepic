```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

## Тепловая матрица корреляций

Матрица корреляции – инструмент, который помогает выявить взаимосвязи между количественными переменными вашего датасета. Построить такую тепловую матрицу можно с помощью **seaborn.heatmap()**.

Но прежде чем визуализировать матрицу корреляций, ее нужно построить. Мы это уже разбирали в позапрошлом модуле, но на всякий случай напомню, что для этого надо сначала **выбрать количественные переменные**, которые будут участвовать в анализе корреляций. Проанализируем признаки 'carat', 'depth', 'table', 'price', 'x', 'y', 'z'.

После того, как вы выбрали признаки, нужно построить матрицу корреляций по датасету с выбранными переменными. Затем можно переходить к визуализации.

```python
corr_matrix = diamonds[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

**Пройдемся по аргументам функции из примера выше:**
* *annot=* – при *True* показывает численное значение корреляции внутри квадрата.
* *cmap=* – задает цветовую палитру *'coolwarm'*, где более холодные оттенки указывают на отрицаительную корреляцию, а теплые – на положительную.
* *vmin=* и *vmax=* – указывают диапазон значений для цветовой шкалы. Без этих аргументов самое маленькое значение было бы самого холодного оттенка, что могло бы запутать нас (или кого-то другого).

Можно настроить тепловую карту, добавив дополнительные параметры, чтобы улучшить читаемость и визуальную привлекательность.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", linewidths=0.5, cmap='coolwarm',
             square=True, cbar_kws={'shrink': .8}, vmax=1, vmin=-1)

plt.title('Матрица корреляции для характеристик бриллиантов')
plt.show()
```

**Что мы здесь поменяли?**
* Округлили значения корреляции до двух знаков после запятой с помощью *fmt='.2f'*.
* Добавили границы между ячейками посредством *linewidths=0.5*.
* Сделали ячейки квадратными при помощи *square=True*.
* *cbar_kws={'shrink' : .8}* уменьшает цветовую шкалу, чтобы она лучше вписывалась в график.
